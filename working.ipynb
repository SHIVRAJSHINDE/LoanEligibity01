```python


listOfColumns = ['RowNumber', 'Loan_ID', 'CustomerId','Surname','Credit_History']

def DropColumns(listOfColumns):
    df.drop(listOfColumns,axis=1,inplace=True)
    return df

df =  DropColumns(listOfColumns)
df
```

```python

def ReplaceWithCorrectValue(colName,valuesToReplace):
    for key,value in valuesToReplace.items():
        df[colName].replace(key,value,inplace=True)
    return df


ReplaceWithCorrectValue(colName='Dependents',valuesToReplace={'3+':3})
ReplaceWithCorrectValue(colName='Loan_Status',valuesToReplace={'Y':1,'N':0})
ReplaceWithCorrectValue(colName='Designation',valuesToReplace={'Assistant Vice President':'AVP','Ma nager': 'Manager'})
ReplaceWithCorrectValue(colName='Married',valuesToReplace={'No':'Single'})
```

```python

def replaceWithMode(columnName):
    for i in columnName:
        modeSelf_Employed = df[i].mode()[0]
        df[i].fillna(modeSelf_Employed,inplace=True)
    return df 

# Replacing the 'Self_Employed','Married' with mode  
columnName = ['Self_Employed','Married','Dependents']
df = replaceWithMode(columnName)

df
```

```python

def replaceWithMean(columnName1):
    for i in columnName1:
        meanOfColumn = df[i].mean()
        df[i].fillna(meanOfColumn,inplace=True)
    return df 

# Replacing the 'Self_Employed','Married' with mode  
columnName1 = ['LoanAmount']
df = replaceWithMean(columnName1)

df
```

```python

import matplotlib.pyplot as plt
import seaborn as sns

def checkOutlier(NumericalFeatures,df):
    plt.figure(figsize=(15, 5))

    for i, feature in enumerate(NumericalFeatures):
        plt.subplot(1, len(NumericalFeatures), i+1)
        sns.boxplot(df[feature])
        plt.xlabel(feature,rotation=45)  # Set x-label as feature name

    plt.ylabel('Values')  # Set y-label

    plt.tight_layout()
    plt.show()


def removeOutlier(colName,df,LowerQuantile,HigherQuantile):
    checkOutlier(NumericalFeatures,df)
    plt.figure(figsize=(10,10))
    sns.boxplot(y =df[colName])

    q1 = df[colName].quantile(LowerQuantile)
    q3 = df[colName].quantile(HigherQuantile)

    IQR = q3-q1

    lowerLimit = q1-1.5*IQR 
    upperLimit = q3+1.5*IQR

    print(q1)
    print(q3)
    print(IQR)
    print(lowerLimit)
    print(upperLimit)
    
    higherOutlier = df[df[colName]>upperLimit]
    df = df.drop(higherOutlier.index,axis=0)
    checkOutlier(NumericalFeatures,df)

    return df

removeOutlier(colName='ApplicantIncome',df=df,LowerQuantile=.20,HigherQuantile=.80)
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import OneHotEncoder
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import OneHotEncoder,OrdinalEncoder
from sklearn.preprocessing import MinMaxScaler,StandardScaler



class SplittingTransformationAndTrainingInitiatorClass:
    def dataReadingAndSplitting(self,final_df):

        y = final_df.loc[:,'Loan_Status']
        X = final_df.drop(['Loan_Status'],axis=1)

        X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.33, random_state = 355)
        return (X_train, X_test, y_train, y_test)

   
    def dataTransformation(self):
        oneHotColumns = ['Self_Employed', 'Geography', 'Property_Area', 'Married', 'Gender']
        trf1 = ColumnTransformer([
            ('OneHot', OneHotEncoder(drop='first', handle_unknown='error'), oneHotColumns)], remainder='passthrough')

        trf2 = ColumnTransformer([
            ('Ordinal1', OrdinalEncoder(categories=[['VP', 'AVP', 'Senior Manager', 'Manager', 'Executive']]), [8])]
            , remainder='passthrough')
        trf3 = ColumnTransformer([
            ('Ordinal2', OrdinalEncoder(categories=[['Graduate', 'Not Graduate']]), [9])]
            , remainder='passthrough')

        trf4 = ColumnTransformer([
            ('scale', StandardScaler(), slice(0, 25))
        ])


        pipe = make_pipeline(trf1,trf2,trf3,trf4)
        return pipe

splittingTransformObj = SplittingTransformationAndTrainingInitiatorClass()
X_train, X_test, y_train, y_test = splittingTransformObj.dataReadingAndSplitting(df)

X_train.isnull().sum()
```

```python
transformationOfData =splittingTransformObj.dataTransformation()

# Call the method on the instance


X_train = transformationOfData.fit_transform(X_train)

```

```python
X_test = transformationOfData.fit_transform(X_test)
```

```python
from sklearn.svm import SVC

modelSVC = SVC()
modelSVC.fit(X_train,y_train)
modelSVC.score(X_train,y_train)
y_predict = modelSVC.predict(X_test)
y_predict


from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_predict)
```
